**Питання 8:** В які години покупці купують більше?

```python
# Конвертація стовпця 'Time' у строку, видалення дробових секунд та конвертація у datetime
df['Time'] = df['Time'].astype(str).str.split('.').str[0]
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time
```

```python
# Витягнення годин з часу
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.hour
```

**Питання 10:** Тестування статистичної гіпотези: чи відрізняється середній розмір покупки в залежності від міста?

- Нульова гіпотеза (H0): Середня сума чеку не відрізняється між чоловіками та жінками.
- Альтернативна гіпотеза (H1): Середня сума чеку відрізняється між чоловіками та жінками.

```python
from statsmodels.stats.weightstats import ztest
```

```python
# Розділення даних за гендером
male_purchases = df[df['Gender'] == 'Male']['Gross Bill Amount']
female_purchases = df[df['Gender'] == 'Female']['Gross Bill Amount']
```

```python
# Виконання z-тесту
z_stat, p_value = ztest(male_purchases, female_purchases)
```

```python
print(f"Z-статистика: {z_stat}")
print(f"P-значення: {p_value}")
```

```python
# Перевірка гіпотези
alpha = 0.05
if p_value < alpha:
    print("Відхиляємо нульову гіпотезу: середня сума чеку відрізняється між чоловіками та жінками.")
else:
    print("Не відхиляємо нульову гіпотезу: середня сума чеку не відрізняється між чоловіками та жінками.")
```

**Висновок:** Оскільки p-значення значно менше за 0.05, ми відхиляємо нульову гіпотезу.

**Рекомендації:**  
Можна розглянути можливість таргетованих маркетингових кампаній, враховуючи різні витрати чоловіків і жінок.
Також слід врахувувати різницю у витратах при формуванні асортименту та пропозицій для кожної статі.
Можна дослідити, які конкретні продукти або категорії товарів можуть бути популярнішими серед жінок і чоловіків для більш детальної оптимізації асортименту і маркетингових зусиль.

```python
# Обчислення кількості покупок для кожного клієнта
customer_purchase_counts = df['Membership ID'].value_counts()

# Обчислення частки повторних покупок
repeat_purchase_ratio = (customer_purchase_counts > 1).mean()

print(f"Частка повторних покупок клієнтів: {repeat_purchase_ratio:.2%}")
```
